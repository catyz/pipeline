```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Gaussian(x, mu, sigma):
    amp = 1/sigma*np.sqrt(2*np.pi)
    exp = np.exp(-(x-mu)**2/2/sigma**2)
    return amp*exp
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, Gaussian(x, 0, 1))
plt.plot(x, Gaussian(x, 0, 5))


```
